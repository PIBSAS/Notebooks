```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from matplotlib.widgets import Slider

#Initial function
x = np.linspace(0, 10,30)
y = np.sin(0.5*x)*np.sin(x*np.random.randn(30))

#Spline Interpolation
spline = UnivariateSpline(x, y, s = 6)
x_spline = np.linspace(0,10,1000)
y_spline  = spline(x_spline)

#Plotting
fig = plt.figure()
ax = fig.subplots()
plt.subplots_adjust(bottom = 0.25) #nos da espacio bajo el gráfico, si es mostrado en ventana.
p = ax.plot(x, y)
p, = ax.plot(x_spline, y_spline, 'g')

#Defining the slider button
ax_slide = plt.axes([0.25, 0.1, 0.65, 0.03]) # xposition, yposition, width, height

#Properties of the Slider
s_factor = Slider(ax_slide, 'Smoothing factor', valmin = 0.1, valmax = 6, valinit= 6, valstep = 0.2) #Object, label, valmin, valmax, valinit, valstep

#Updating the plot
def update(val):
    current_v = s_factor.val
    spline = UnivariateSpline(x, y, s = current_v)
    p.set_ydata(spline(x_spline))
    fig.canvas.draw() #redraw the figure

#Calling the function "update" when the Slider is changed
s_factor.on_changed(update)

plt.show()
```

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Initial data
x = np.linspace(0, 10, 30)
m_true = 0.5  # Pendiente verdadera
b_true = 0.2  # Intercepto verdadero
y_true = m_true * x + b_true

# Función para calcular la línea con la pendiente y el intercepto dados
def calculate_line(m, b):
    return m * x + b

# Plotting
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

# Dibujar la línea inicial
line, = ax.plot(x, calculate_line(m_true, b_true), 'r', linewidth=2)

# Definir el slider para la pendiente (m) y el intercepto (b)
ax_m = plt.axes([0.25, 0.15, 0.65, 0.03])
ax_b = plt.axes([0.25, 0.1, 0.65, 0.03])

slider_m = Slider(ax_m, 'Pendiente (m)', valmin=-1, valmax=1, valinit=m_true, valstep=0.01)
slider_b = Slider(ax_b, 'Intercepto (b)', valmin=-1, valmax=1, valinit=b_true, valstep=0.01)

# Función de actualización para el cambio de pendiente e intercepto
def update(val):
    m = slider_m.val
    b = slider_b.val
    line.set_ydata(calculate_line(m, b))
    fig.canvas.draw_idle()

slider_m.on_changed(update)
slider_b.on_changed(update)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Initial data
x = np.linspace(-10, 10, 30)
a = 2  # Pendiente inicial
b = 1  # Intercepto inicial
c = 2
# Función para calcular la línea con la pendiente y el intercepto dados
def calculate_square(a,b, c):
    return a*x**2 + b*x + c

# Plotting
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.35)

# Dibujar la línea inicial
line, = ax.plot(x, calculate_square(a, b, c), 'r', linewidth=2)

# Definir el slider para la pendiente (m) y el corte de ordenada (b)
ax_a = plt.axes([0.25, 0.20, 0.65, 0.03])
ax_b = plt.axes([0.25, 0.15, 0.65, 0.03])
ax_c = plt.axes([0.25, 0.10, 0.65, 0.03])

slider_a = Slider(ax_a, 'Coheficiente Principal (a)', valmin=-5, valmax=5, valinit=a, valstep=0.01)
slider_b = Slider(ax_b, 'Coheficiente Lineal (b)', valmin=-5, valmax=5, valinit=b, valstep=0.01)
slider_c = Slider(ax_c, 'Corte de Ordenada (c)', valmin = -5, valmax = 5, valinit = c, valstep = 0.01)

# Función de actualización para el cambio de pendiente e intercepto
def update(val):
    a = slider_a.val
    b = slider_b.val
    c = slider_c.val
    line.set_ydata(calculate_square(a, b, c))
    fig.canvas.draw_idle()

slider_a.on_changed(update)
slider_b.on_changed(update)
slider_c.on_changed(update)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython import display as dp
```

```python
def line(m, b):
    x = np.linspace(-10, 10, 100)
    y = m*x +b
    plt.plot(x, y)
    plt.axis("scaled")
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.show()
```

```python
line(1,5)
```

## Sliders

```python
slider1 = widgets.IntSlider(min=0, max=5, value=1, continuous_update=False, description="Pendiente")
slider2 = widgets.FloatSlider(min=1, max=5, value=0, step=0.1, description="Corte de ordenada")

gui = widgets.interactive(line, m=slider1, b=slider2)
dp.display(gui)
```

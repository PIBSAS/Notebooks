# **La suma de Riemann**
Es el nombre que recibe el cálculo aproximado de una integral definida, mediante una sumatoria discreta con un número de términos finito. Una aplicación común es la aproximación del área de funciones en un gráfico.

Fue el matemático alemán Georg Friedrich Bernhard Riemann (1826-1866) quien ofreció por primera vez una definición rigurosa de la integral de una función en un intervalo dado. La dio a conocer en un artículo publicado en 1854.  

<br>  

Suma-de-Riemann-01.jpg

La suma de Riemann se define sobre una función $y = f(x)$, con $x$ perteneciente al intervalo cerrado [a, b]. Sobre este intervalo se efectúa una partición P de n elementos:

$$P = \lbrace{x_0=a, x_1, x_2, ..., x_n=b \rbrace}$$

Esto significa que el intervalo se divide de la siguiente manera:

$$S(f, n) = \sum_{k = 1}^{n} f(t_k)(x_k - x_{k-1}$$
<br>
Aquí $t_k$ está comprendido entre $x_{k-1}$ y $x_k$: $x_{k-1} \leq t_k \leq x_k$

En la figura 1 se muestra gráficamente la suma de Riemann de la función f en el intervalo $[x_0, x_4]$ sobre una partición de cuatro subintervalos, los rectángulos de color gris.

La suma representa el área total de los rectángulos y el resultado de esta suma se aproxima numéricamente al área bajo la curva $f$, entre las abscisas $x=x_0$ y $x=x_4$.

Desde luego, la aproximación al área bajo la curva mejora muchísimo en la medida que el número n de particiones sea mayor. De esta manera la suma converge al área bajo la curva, cuando el número $n$ de particiones tiende a infinito.

# Fórmulas y propiedades

La suma de Riemann de la función f(x) sobre la partición:

$$P = \lbrace{x_0=a, x_1, x_2, ..., x_n=b \rbrace}$$

Definida sobre el intervalo $[a, b]$, está dada por:

$$S(P, f) = \sum_{k = 1}^{n} f(t_k)(x_k - x_{k-1}$$

Donde $tk$ es un valor en el intervalo $[x_k, x_{k-1}]$. En la suma de Riemann se suelen usar intervalos regulares de ancho $\Delta x = \frac{\left( b - a \right)}{n}$, donde $a$ y $b$ son los valores mínimo y máximo de la abscisa, mientras que $n$ es el número de subdivisiones.

En ese caso la suma derecha de Riemann es:

$$Sd(f, n) = [f(a+ \Delta x) + f(a + 2 \Delta x) + ... + f(a + (n-1) \Delta x) + f(b)] * \Delta x$$
<br>

Mientras que la suma izquierda de Riemann se expresa como:

$$Si(f, n) = [f(a) + f(a + \Delta x) + ... + f(a + (n-1) \Delta x)] \cdot \Delta x$$

Suma inferior queda definida como:

$$S_{inf}(f, P) = \sum_{k=1}^{n} m_k \cdot (x_k - x_{k-1})$$

Donde en cada subintervalo $[x_{k-1}, x_k]$, se considera el *mínimo valor* que toma la funcion $f(x)$:

$$m_{k} = \min_{x \in \left[x_{k-1}, x_k\right]} f(x)$$

Suma superior queda definida como:

$$S_{sup}(f, P) = \sum_{k=1}^{n} M_k \cdot (x_k - x_{k-1})$$

Donde en cada subintervalo $[x_{k-1}, x_k]$, se considera el *máximo valor* que toma la funcion $f(x)$:

$$M_{k} = \max_{x \in \left[x_{k-1}, x_k\right]} f(x)$$

```python
import math

entradas = {
    "sen": "math.sin",
    "sin": "math.sin",
    "cos": "math.cos",
    "tan": "math.tan",
    "e^": "math.exp",
    "ln": "math.log",
    "log": "math.log10",
    "x^": "x**"
}

ingreso = ""

def sanitizar(sintaxis):
    for clave, valor in entradas.items():
        sintaxis = sintaxis.replace(clave, valor)
    return sintaxis

def sumaInferior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")
    
    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria_I = 0.0

    for k in range(n):
        izq = a + k * delta
        der = izq + delta

        m_k = min(funcion(izq), funcion(der))
        
        sumatoria_I += m_k * delta

    return sumatoria_I

def sumaSuperior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")
    
    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria_S = 0.0

    for k in range(n):
        izq = a + k * delta
        der = izq + delta

        M_k = max(funcion(izq), funcion(der))
        
        sumatoria_S += M_k * delta

    return sumatoria_S

def SumaSupSumaInf(funcion, minimo, maximo, eps):
    global ingreso
    N = 1
    while True:
        inf = sumaInferior(funcion, minimo, maximo, N)
        sup = sumaSuperior(funcion, minimo, maximo, N)
        diferencia = abs(sup - inf)
        print(f"{N:3d}: {inf:.6f} < Integral < {sup:.6f}")
        if diferencia < eps:
            print(f"\n→ Con N = {N}, la diferencia {diferencia:.6f} < ε = {eps}")
            break
        N += 1

def main():
    global ingreso
    print("Aproximación de la integral con suma superior e inferior")

    ingreso = input("Ingrese la función f(x): ")
    f = sanitizar(ingreso)
    funcion = lambda x: eval(f, {"x": x, "math": math})

    minimo = float(input("Desde (a): "))
    maximo = float(input("Hasta (b): "))
    eps = float(input("Epsilon (tolerancia, ej: 0.001): "))

    print("\nSuma inferior y superior por número de rectángulos:\n")
    SumaSupSumaInf(funcion, minimo, maximo, eps)

main()
```

```python
import math

from dash import Dash, dcc, html, dash_table
from dash.dependencies import Input, Output, State
import dash_bootstrap_components as dbc
import numpy as np
import plotly.graph_objects as go

app = Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

entradas = {
    "sen": "math.sin",
    "sin": "math.sin",
    "cos": "math.cos",
    "tan": "math.tan",
    "e^": "math.exp",
    "ln": "math.log",
    "log": "math.log10",
    "x^": "x**"
}

def sanitizar(sintaxis):
    for clave, valor in entradas.items():
        sintaxis = sintaxis.replace(clave, valor)
    return sintaxis

def sumaInferior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")
    
    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria_I = 0.0

    for k in range(n):
        izq = a + k * delta
        der = izq + delta

        m_k = min(funcion(izq), funcion(der))
        
        sumatoria_I += m_k * delta

    return sumatoria_I

def sumaSuperior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")
    
    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria_S = 0.0

    for k in range(n):
        izq = a + k * delta
        der = izq + delta

        M_k = max(funcion(izq), funcion(der))
        
        sumatoria_S += M_k * delta

    return sumatoria_S

def SumaSupSumaInf(funcion, minimo, maximo, eps):
    global ingreso
    N = 1
    while True:
        inf = sumaInferior(funcion, minimo, maximo, N)
        sup = sumaSuperior(funcion, minimo, maximo, N)
        diferencia = abs(sup - inf)
        print(f"{N:3d}: {inf:.6f} < Integral < {sup:.6f}")
        if diferencia < eps:
            print(f"\n→ Con N = {N}, la diferencia {diferencia:.6f} < ε = {eps}")
            break
        N += 1

def amain():
    global ingreso
    print("Aproximación de la integral con suma superior e inferior")

    ingreso = input("Ingrese la función f(x): ")
    f = sanitizar(ingreso)
    funcion = lambda x: eval(f, {"x": x, "math": math})

    minimo = float(input("Desde (a): "))
    maximo = float(input("Hasta (b): "))
    eps = float(input("Epsilon (tolerancia, ej: 0.001): "))

    print("\nSuma inferior y superior por número de rectángulos:\n")
    SumaSupSumaInf(funcion, minimo, maximo, eps)

#main()

app.layout = dbc.Container([
    html.H2("Suma inferior y superior de Riemann", className="text-center my-4"),
    
    html.Div([
        dbc.Row([
            dbc.Col([
                dbc.Label("Función f(x)"),
                dbc.Input(id="funcion-input", placeholder="Ej: x**2, math.sin(x), math.exp(x)", type="text"),
            ], width=3),
            dbc.Col([
                dbc.Label("Desde (a)"),
                dbc.Input(id="desde-input", placeholder="Desde a", type="number", value=0),
            ], width=2),
            dbc.Col([
                dbc.Label("Hasta (b)"),
                dbc.Input(id="hasta-input", placeholder="Hasta b", type="number", value=1),
            ], width=2),
            dbc.Col([
                dbc.Label("Epsilon (ε)"),
                dbc.Input(id="eps-input", placeholder="Epsilon", type="number", value=0.01, step=0.001),
            ], width=2),
            dbc.Col([
                dbc.Label(" "),
                dbc.Button("Calcular", id="start-btn", color="primary", className="w-100"),
            ], width=3),
        ], className="align-items-end"),
    ], className="mb-4 p-3 bg-light rounded"),
    
    dbc.Alert(id="error-alert", color="danger", is_open=False, dismissable=True),
    
    html.H4("Tabla de Resultados", className="my-3"),
    dash_table.DataTable(
        id='tabla-riemann',
        columns=[
            {"name": "N", "id": "N"},
            {"name": "Suma Inferior", "id": "SumaInferior"},
            {"name": "Suma Superior", "id": "SumaSuperior"},
            {"name": "Diferencia", "id": "Diferencia"}
        ],
        data=[],
        style_cell={'textAlign': 'center'},
        style_header={'fontWeight': 'bold'},
        page_size=10
    ),
    
    html.H4("Gráfico de la Función y Rectángulos", className="my-3"),
    dcc.Graph(id="grafico-riemann"),
], fluid=True)
@app.callback(
    Output('tabla-riemann', 'data'),
    Output('grafico-riemann', 'figure'),
    Output('error-alert', 'children'),
    Output('error-alert', 'is_open'),
    Input('start-btn', 'n_clicks'),
    State('funcion-input', 'value'),
    State('desde-input', 'value'),
    State('hasta-input', 'value'),
    State('eps-input', 'value')
)
def calcular(n_clicks, funcion_texto, desde, hasta, eps):
    if not n_clicks:
        return [], go.Figure(), "", False
    
    # Validar entradas
    if not funcion_texto:
        return [], go.Figure(), "Por favor ingrese una función", True
    
    if desde is None or hasta is None:
        return [], go.Figure(), "Por favor ingrese los límites de integración", True
    
    if eps is None or eps <= 0:
        return [], go.Figure(), "Epsilon debe ser mayor a cero", True
    
    if hasta <= desde:
        return [], go.Figure(), "El límite superior debe ser mayor al inferior", True
    
    try:
        # Sanitizar y crear la función
        f_python = sanitizar(funcion_texto)
        # Asegurarse de que la función sea evaluable
        test_x = (desde + hasta) / 2
        test_result = eval(f_python, {"x": test_x, "math": math})
        if not isinstance(test_result, (int, float)):
            return [], go.Figure(), "La función debe devolver un valor numérico", True
        
        f = lambda x: eval(f_python, {"x": x, "math": math})
        
        N = 1
        datos = []
        max_iteraciones = 1000  # Límite para evitar bucles infinitos
        
        while N <= max_iteraciones:
            try:
                inf = sumaInferior(f, desde, hasta, N)
                sup = sumaSuperior(f, desde, hasta, N)
                diferencia = abs(sup - inf)
                
                datos.append({
                    "N": N,
                    "SumaInferior": round(inf, 6),
                    "SumaSuperior": round(sup, 6),
                    "Diferencia": round(diferencia, 6)
                })
                
                if diferencia < eps:
                    break
                N += 1
            except Exception as e:
                return [], go.Figure(), f"Error en cálculo: {str(e)}", True
        
        # Crear gráfico
        x_cont = np.linspace(desde, hasta, 500)
        y_cont = [f(x) for x in x_cont]
        fig = go.Figure()
        
        # Graficar la función
        fig.add_trace(go.Scatter(
            x=x_cont, 
            y=y_cont, 
            mode='lines', 
            name=f'f(x) = {funcion_texto}',
            line=dict(color='black', width=2)
        ))
        
        # Agregar rectángulos
        dx = (hasta - desde) / N
        x_rects = [desde + k * dx for k in range(N)]
        
        for i, x0 in enumerate(x_rects):
            x1 = x0 + dx
            h_inf = min(f(x0), f(x1))
            h_sup = max(f(x0), f(x1))
            
            # Rectángulo inferior
            fig.add_trace(go.Scatter(
                x=[x0, x0, x1, x1, x0],
                y=[0, h_inf, h_inf, 0, 0],
                fill="toself",
                fillcolor="blue",
                line=dict(color="blue"),
                opacity=0.3,
                showlegend=(i == 0),
                name="Suma Inferior"
            ))
            
            # Rectángulo superior
            fig.add_trace(go.Scatter(
                x=[x0, x0, x1, x1, x0],
                y=[0, h_sup, h_sup, 0, 0],
                fill="toself",
                fillcolor="red",
                line=dict(color="red"),
                opacity=0.3,
                showlegend=(i == 0),
                name="Suma Superior"
            ))
        
        fig.update_layout(
            title=f"Suma inferior (azul) y superior (roja) de Riemann - N={N}",
            xaxis_title="x",
            yaxis_title="f(x)",
            showlegend=True,
            height=500
        )
        
        return datos, fig, "", False
        
    except Exception as e:
        return [], go.Figure(), f"Error: {str(e)}", True

if __name__ == "__main__":
    app.run(debug=True, port=8099)
```

```python
import math
from dash import Dash, dcc, html, dash_table, callback_context
from dash.dependencies import Input, Output, State
import dash_bootstrap_components as dbc
import numpy as np
import plotly.graph_objects as go
import time

app = Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

entradas = {
    "sen": "math.sin",
    "sin": "math.sin",
    "cos": "math.cos",
    "tan": "math.tan",
    "e^": "math.exp",
    "ln": "math.log",
    "log": "math.log10",
    "x^": "x**"
}

def sanitizar(sintaxis):
    for clave, valor in entradas.items():
        sintaxis = sintaxis.replace(clave, valor)
    return sintaxis

def sumaInferior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")
    
    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria_I = 0.0

    for k in range(n):
        izq = a + k * delta
        der = izq + delta
        m_k = min(funcion(izq), funcion(der))
        sumatoria_I += m_k * delta

    return sumatoria_I

def sumaSuperior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")
    
    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria_S = 0.0

    for k in range(n):
        izq = a + k * delta
        der = izq + delta
        M_k = max(funcion(izq), funcion(der))
        sumatoria_S += M_k * delta

    return sumatoria_S

def SumaSupSumaInf(funcion, minimo, maximo, eps):
    N = 1
    resultados = []
    while True:
        inf = sumaInferior(funcion, minimo, maximo, N)
        sup = sumaSuperior(funcion, minimo, maximo, N)
        diferencia = abs(sup - inf)
        
        resultados.append({
            'N': N,
            'SumaInferior': inf,
            'SumaSuperior': sup,
            'Diferencia': diferencia
        })
        
        if diferencia < eps:
            break
        N += 1
        
        # Limitar iteraciones para no bloquear la interfaz
        if N > 10000:
            break
            
    return resultados

# Variable global para almacenar el estado del cálculo
calculo_estado = {
    'en_progreso': False,
    'datos': [],
    'funcion_texto': None,
    'f_python': None,
    'desde': None,
    'hasta': None,
    'eps': None,
    'completado': False,
    'N_final': 0,
    'resultados_crudos': []
}

app.layout = dbc.Container([
    html.H2("Suma inferior y superior de Riemann", className="text-center my-4"),
    
    html.Div([
        dbc.Row([
            dbc.Col([
                dbc.Label("Función f(x)"),
                dbc.Input(id="funcion-input", placeholder="Ej: x**2, sin(x), e^x", type="text", value="x**2"),
            ], width=3),
            dbc.Col([
                dbc.Label("Desde (a)"),
                dbc.Input(id="desde-input", placeholder="Desde a", type="number", value=0),
            ], width=2),
            dbc.Col([
                dbc.Label("Hasta (b)"),
                dbc.Input(id="hasta-input", placeholder="Hasta b", type="number", value=1),
            ], width=2),
            dbc.Col([
                dbc.Label("Epsilon (ε)"),
                dbc.Input(id="eps-input", placeholder="Epsilon", type="number", value=0.1, step=0.001),
            ], width=2),
            dbc.Col([
                dbc.Label(" "),
                dbc.Button("Calcular", id="start-btn", color="primary", className="w-100"),
                dbc.Button("Detener", id="stop-btn", color="danger", className="w-100 mt-2", disabled=True),
            ], width=3),
        ], className="align-items-end"),
    ], className="mb-4 p-3 bg-light rounded"),
    
    dbc.Alert(id="error-alert", color="danger", is_open=False, dismissable=True),
    dbc.Alert(id="info-alert", color="info", is_open=False, dismissable=True),
    
    dbc.Progress(id="progress-bar", value=0, striped=True, animated=True, className="mb-3"),
    
    html.H4("Tabla de Resultados", className="my-3"),
    dash_table.DataTable(
        id='tabla-riemann',
        columns=[
            {"name": "N", "id": "N"},
            {"name": "Suma Inferior", "id": "SumaInferior"},
            {"name": "Integral", "id": "Integral"},
            {"name": "Suma Superior", "id": "SumaSuperior"},
            {"name": "Diferencia", "id": "Diferencia"}
        ],
        data=[],
        style_cell={'textAlign': 'center'},
        style_header={'fontWeight': 'bold'},
        page_size=10,
        style_data_conditional=[
            {
                'if': {'row_index': 'odd'},
                'backgroundColor': 'rgb(248, 248, 248)'
            },
            {
                'if': {'column_id': 'Integral'},
                'fontWeight': 'bold',
                'color': 'darkblue'
            }
        ]
    ),
    
    html.H4("Gráfico de la Función y Rectángulos", className="my-3"),
    dcc.Graph(id="grafico-riemann"),
    
    # Componente oculto para trigger de actualizaciones
    dcc.Interval(id="interval-update", interval=100, n_intervals=0, disabled=True),
    
    # Almacenamiento para el estado del cálculo
    dcc.Store(id='calculation-store', data=calculo_estado),
], fluid=True)

@app.callback(
    Output('calculation-store', 'data'),
    Output('start-btn', 'disabled'),
    Output('stop-btn', 'disabled'),
    Output('interval-update', 'disabled'),
    Output('info-alert', 'children'),
    Output('info-alert', 'is_open'),
    Output('error-alert', 'children'),
    Output('error-alert', 'is_open'),
    Input('start-btn', 'n_clicks'),
    Input('stop-btn', 'n_clicks'),
    Input('interval-update', 'n_intervals'),
    State('funcion-input', 'value'),
    State('desde-input', 'value'),
    State('hasta-input', 'value'),
    State('eps-input', 'value'),
    State('calculation-store', 'data')
)
def control_calculo(start_clicks, stop_clicks, n_intervals, funcion_texto, desde, hasta, eps, store_data):
    ctx = callback_context
    if not ctx.triggered:
        return store_data, False, True, True, "", False, "", False
    
    trigger_id = ctx.triggered[0]['prop_id'].split('.')[0]
    
    if trigger_id == 'start-btn' and start_clicks:
        # Validar entradas
        if not funcion_texto:
            return store_data, False, True, True, "", False, "Por favor ingrese una función", True
        
        if desde is None or hasta is None:
            return store_data, False, True, True, "", False, "Por favor ingrese los límites de integración", True
        
        if eps is None or eps <= 0:
            return store_data, False, True, True, "", False, "Epsilon debe ser mayor a cero", True
        
        if hasta <= desde:
            return store_data, False, True, True, "", False, "El límite superior debe ser mayor al inferior", True
        
        try:
            # Sanitizar y crear la función
            f_python = sanitizar(funcion_texto)
            test_x = (desde + hasta) / 2
            test_result = eval(f_python, {"x": test_x, "math": math, "sin": math.sin, "cos": math.cos, "tan": math.tan, "exp": math.exp, "log": math.log, "log10": math.log10})
            
            # Inicializar estado del cálculo
            nuevo_estado = {
                'en_progreso': True,
                'datos': [],
                'funcion_texto': funcion_texto,
                'f_python': f_python,
                'desde': desde,
                'hasta': hasta,
                'eps': eps,
                'completado': False,
                'N_final': 0,
                'resultados_crudos': [],
                'indice_actual': 0,
                'grafico_generado': False
            }
            return nuevo_estado, True, False, False, "Calculando...", True, "", False
            
        except Exception as e:
            return store_data, False, True, True, "", False, f"Error: {str(e)}", True
    
    elif trigger_id == 'stop-btn' and stop_clicks:
        # Detener cálculo
        store_data['en_progreso'] = False
        return store_data, False, True, True, "Cálculo detenido", True, "", False
    
    elif trigger_id == 'interval-update' and store_data.get('en_progreso', False):
        # Ejecutar cálculo por lotes para mejor rendimiento
        if not store_data.get('resultados_crudos'):
            try:
                # Ejecutar tu función SumaSupSumaInf una sola vez
                f = lambda x: eval(store_data['f_python'], {"x": x, "math": math, "sin": math.sin, "cos": math.cos, "tan": math.tan, "exp": math.exp, "log": math.log, "log10": math.log10})
                store_data['resultados_crudos'] = SumaSupSumaInf(f, store_data['desde'], store_data['hasta'], store_data['eps'])
                store_data['N_final'] = store_data['resultados_crudos'][-1]['N'] if store_data['resultados_crudos'] else 0
            
            except Exception as e:
                store_data['en_progreso'] = False
                return store_data, False, True, True, "", False, f"Error en cálculo: {str(e)}", True
        
        # Procesar resultados en lotes para mostrar progreso
        indice_actual = store_data.get('indice_actual', 0)
        resultados_crudos = store_data.get('resultados_crudos', [])
        lote_size = min(50, len(resultados_crudos) - indice_actual)  # Procesar hasta 50 resultados por intervalo
        
        if indice_actual < len(resultados_crudos):
            # Procesar lote actual
            for i in range(indice_actual, min(indice_actual + lote_size, len(resultados_crudos))):
                resultado = resultados_crudos[i]
                store_data['datos'].append({
                    "N": resultado['N'],
                    "SumaInferior": f"{resultado['SumaInferior']:.6f}",
                    "Integral": f"∫{store_data['funcion_texto']} dx",
                    "SumaSuperior": f"{resultado['SumaSuperior']:.6f}",
                    "Diferencia": f"{resultado['Diferencia']:.6f}"
                })
            
            store_data['indice_actual'] = indice_actual + lote_size
            
            # Actualizar progreso
            progreso_porcentaje = (store_data['indice_actual'] / len(resultados_crudos)) * 100
            mensaje = f"Procesando... {store_data['indice_actual']}/{len(resultados_crudos)} iteraciones ({progreso_porcentaje:.1f}%)"
            
            return store_data, True, False, False, mensaje, True, "", False
        
        else:
            # Cálculo completado
            store_data['en_progreso'] = False
            store_data['completado'] = True
            store_data['grafico_generado'] = True  # Marcar para generar gráfico
            mensaje = f"¡Cálculo completado! N final: {store_data['N_final']}, Iteraciones: {len(resultados_crudos)}"
            return store_data, False, True, True, mensaje, True, "", False
    
    return store_data, False, True, True, "", False, "", False

@app.callback(
    Output('tabla-riemann', 'data'),
    Output('grafico-riemann', 'figure'),
    Output('progress-bar', 'value'),
    Input('interval-update', 'n_intervals'),
    State('calculation-store', 'data')
)
def actualizar_ui(n_intervals, store_data):
    if not store_data.get('datos'):
        return [], go.Figure(), 0
    
    datos = store_data['datos']
    
    # Calcular progreso
    if store_data.get('completado', False):
        progreso = 100
    elif store_data.get('resultados_crudos'):
        resultados_crudos = store_data.get('resultados_crudos', [])
        indice_actual = store_data.get('indice_actual', 0)
        progreso = (indice_actual / len(resultados_crudos)) * 100 if resultados_crudos else 0
    else:
        progreso = 0
    
    # Crear gráfico - AHORA SIEMPRE QUE HAYA DATOS
    fig = go.Figure()
    
    if datos and store_data.get('grafico_generado', False):
        try:
            f = lambda x: eval(store_data['f_python'], {"x": x, "math": math, "sin": math.sin, "cos": math.cos, "tan": math.tan, "exp": math.exp, "log": math.log, "log10": math.log10})
            
            # Graficar función
            x_cont = np.linspace(store_data['desde'], store_data['hasta'], 500)
            y_cont = [f(x) for x in x_cont]
            fig.add_trace(go.Scatter(
                x=x_cont, y=y_cont, mode='lines', 
                name=f'f(x) = {store_data["funcion_texto"]}',
                line=dict(color='black', width=2)
            ))
            
            # Agregar rectángulos (solo si N es razonable)
            N_final = store_data.get('N_final', 0)
            if N_final <= 100:  # Aumentado el límite para mejor visualización
                dx = (store_data['hasta'] - store_data['desde']) / N_final
                x_rects = [store_data['desde'] + k * dx for k in range(N_final)]
                
                # Solo agregar algunos rectángulos para no sobrecargar (cada 5 por ejemplo)
                for i in range(0, min(N_final, 20)):  # Máximo 20 rectángulos para no saturar
                    x0 = x_rects[i]
                    x1 = x0 + dx
                    h_inf = min(f(x0), f(x1))
                    h_sup = max(f(x0), f(x1))
                    
                    # Rectángulo inferior
                    fig.add_trace(go.Scatter(
                        x=[x0, x0, x1, x1, x0], y=[0, h_inf, h_inf, 0, 0],
                        fill="toself", fillcolor="blue", line=dict(color="blue"),
                        opacity=0.3, showlegend=(i == 0), name="Suma Inferior"
                    ))
                    
                    # Rectángulo superior
                    fig.add_trace(go.Scatter(
                        x=[x0, x0, x1, x1, x0], y=[0, h_sup, h_sup, 0, 0],
                        fill="toself", fillcolor="red", line=dict(color="red"),
                        opacity=0.3, showlegend=(i == 0), name="Suma Superior"
                    ))
            
            fig.update_layout(
                title=f"Suma inferior (azul) y superior (roja) de Riemann - N={N_final}",
                xaxis_title="x", yaxis_title="f(x)", showlegend=True, height=500
            )
            
        except Exception as e:
            print(f"Error al crear gráfico: {str(e)}")
            # Si hay error, al menos mostrar la función
            fig.add_annotation(
                text=f"Error al generar rectángulos: {str(e)}",
                xref="paper", yref="paper",
                x=0.5, y=0.5, showarrow=False,
                font=dict(size=16, color="red")
            )
    
    return datos, fig, progreso

if __name__ == "__main__":
    app.run(debug=True, port=8099)
```
